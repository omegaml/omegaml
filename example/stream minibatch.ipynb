```python
from multiprocessing import Process
from omegastream.minibatch import stream

db = om.datasets.mongodb

# fixed size -- emits windows of fixed sizes
@stream('test', size=2, keep=True)
def myprocess(window):
    try:
        db.processed.insert_one({'data': window.data or {}})
        print(window.data)
    except Exception as e:
        print(e)
    return window
```

```python
# relaxed time windows -- emits all documents within most recent interval
from multiprocessing import Process
from omegastream.minibatch import stream

db = om.datasets.mongodb

@stream('test', interval=1, keep=True)
def myprocess(window):
    print("ok")
    try:
        print(window.data)
        db.processed.insert_one({'data': window.data or {}})
    except Exception as e:
        print(e)
    return window
```

```python
# fixed time windows -- emits documents that were inserted in exactly a given time interval
from multiprocessing import Process
from omegastream.minibatch import stream
import datetime 

db = om.datasets.mongodb

@stream('test', interval=2, relaxed=False, keep=True)
def myprocess(window):
    print(datetime.datetime.now())
    try:
        print(window.data)
        db.processed.insert_one({'data': window.data or {}})
    except Exception as e:
        print(e)
    return window
```
