```r
# load omegaml
library(reticulate)
om <- import("omegaml")
```

```r
# a sample r model
library(caret)
data(mtcars)
model <- train(mpg ~ wt,
               data = mtcars,
              method = "lm")
predict(model, mtcars)
om$models$put('r$model', 'mtcars-model')
om$datasets$put(mtcars, 'mtcars')
```

```r
# run prediction on runtime
result <- om$runtime$model('mtcars-model')$predict(mtcars)
result$get()
```
